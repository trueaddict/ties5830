```python
from desdeo_problem.testproblems.TestProblems import test_problem_builder
from desdeo_emo.EAs import RVEA, NSGAIII
import matplotlib.pyplot as plt
import numpy as np
```

```python
# RVEA
# A) Degenerate -> DTLZ5
# n_of_objectives=5, n_of_variables=10

# I reduced the number of iteration to 200 because it took very long to process with 1000 iterations

objectives = [2, 5]
variables = 10
test_problem = 'DTLZ5'

for n in objectives:
    dtlz5 = test_problem_builder('DTLZ5', n_of_objectives=n, n_of_variables=10)
    evolver = RVEA(dtlz5, n_gen_per_iter=200, n_iterations=20)
    while evolver.continue_evolution():
        evolver.iterate()
    individuals, solutions = evolver.end()
    # Not sure if I am plotting right, but I tried
    fig = plt.figure()
    ax = fig.add_subplot(title="A) Degenerate RVEA")
    ax.scatter(x=individuals[:,0], y=individuals[:,1])
```

```python

# NSGAIII
# A) Degenerate -> DTLZ5

objectives = [2, 5]
variables = 10
test_problem = 'DTLZ5'

for n in objectives:
    dtlz5 = test_problem_builder(test_problem, n_of_objectives=n, n_of_variables=10)
    evolver = NSGAIII(dtlz5, n_gen_per_iter=200, n_iterations=20)
    while evolver.continue_evolution():
        evolver.iterate()
    individuals, solutions = evolver.end()
    fig = plt.figure()
    ax = fig.add_subplot(title="A) Degenerate NSGAIII")
    ax.scatter(x=individuals[:,0], y=individuals[:,1])
```

```python
# NSGAIII
# n_of_objectives=5, n_of_variables=10

dtlz5 = test_problem_builder('DTLZ5', n_of_objectives=5, n_of_variables=10)
evolver = NSGAIII(dtlz5, n_gen_per_iter=200, n_iterations=20)
while evolver.continue_evolution():
    evolver.iterate()
individuals, solutions = evolver.end()

plt.scatter(x=individuals[:,0], y=individuals[:,1])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.view_init(30,45)
ax.scatter(solutions[:,0], solutions[:,1], solutions[:,2], label="NSGAIII")

ax.scatter(individuals[:,0], individuals[:,1], individuals[:,2], label="NSGAIII")
```

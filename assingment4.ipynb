```python
from desdeo_problem.testproblems.TestProblems import test_problem_builder

DTLZ2 = test_problem_builder('DTLZ2', n_of_objectives=3, n_of_variables=5)

from scipy.stats import qmc
sampler = qmc.LatinHypercube(d=5)
sample_data = sampler.random(n=50)

print('sample_data', sample_data)

DTLZ2_obj_val = DTLZ2.evaluate(sample_data)

print('DTLZ2_obj_val.objectives', DTLZ2_obj_val.objectives)
```

```python
from desdeo_problem.surrogatemodels import GaussianProcessRegressor

model = GaussianProcessRegressor()

model.fit(X=sample_data, y = DTLZ2_obj_val.objectives)

```

```python
test_data = sampler.random(n=20)

print(test_data)

y_predict, y_delta = model.predict(test_data)

import matplotlib.pyplot as plt

print(y_predict)
```
